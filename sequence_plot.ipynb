```python
import seq_driver 
seq_length = 10
```

```python
seq_dict = seq_driver.seq_driver_incr(10,10)
```

```python
import matplotlib.pyplot as plt
```

```python
for key, seq in seq_dict.items(): 
         plt.plot(range(seq_length), seq, label=str(key))
plt.ylabel('Fibonacci-like sequence')
plt.xlabel('sequence index')

plt.legend()

plt.show()
```

```python
import numpy as np
import mpl_toolkits.mplot3d.axes3d as p3
```

```python
fig = plt.figure() 
ax = p3.Axes3D(fig)

for key, seq in seq_dict.items():
    ax.plot(key[0]*np.ones(seq_length), range(seq_length), seq, label=str(key)) 
    
ax.legend()
```

```python
initial = [key[0] for key in seq_dict]
X,Y = np.meshgrid(range(seq_length), initial)
Z = np.ndarray((len(initial), seq_length))
for key, seq in seq_dict.items(): 
    Z[key[0]-2,:] = seq
```

```python
fig = plt.figure() 
ax = p3.Axes3D(fig) 

ax.plot_surface(X,Y,Z)
```
